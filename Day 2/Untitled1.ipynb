```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,2*np.pi)
y = np.sin(x)
y1 = np.cos(x)
plt.plot(x,y)
plt.plot(x,y1)
plt.show()
```

```python
x =[1,2,3,4,5,6,7,8]
y = [5,3,1,2,6,1,4,6]
plt.scatter(x,y)
plt.show()
```

```python
x = np.linspace(0,2,20)
y = x**2
y1 = x
plt.plot(x,y1,'o-',label ='linear')
plt.plot(x,y,'x-',label ='quadratic')
plt.legend(loc = 'best')
plt.title('Linear And  Quadratic Waveforms')
plt.xlabel('Time')
plt.ylabel('Resultant Waveforms')
plt.show()
```

```python
fig = plt.figure()
axis = fig.add_axes([0,0,1,1])
dishes = ['jalebi','dosa','chole-bathure','naan-daal','chowmein']
people =[89,76,43,12,87]
axis.bar(dishes,people)
plt.title('Dishes Favourites',size = 20)
plt.xlabel('Names of Dishes')
plt.ylabel('Population who eats')
plt.show()
```

```python
dishes = ['jalebi','dosa','chole-bathure','naan-daal']
colors = ['red','blue','green','yellow']
angles =[38.4,40.6,20.7,10.3]
patches,text = plt.pie(angles,colors = colors, shadow = False,startangle =90)
plt.legend(patches,dishes,loc ="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('student-mat.csv')
df.plot(kind = 'hist', bins =[0,5,7,10,15,20,25,35],rwidth =0.8)
plt.show()
```

```python
df.plot(kind = 'density',subplots =True,sharex =False,sharey =False,layout =(4,4))
plt.show()
```

```python
df['absences'].plot(kind = 'density')
plt.show()
```

```python
df['absences'].plot(kind = 'box')
plt.show()
```

```python
df2 = df.groupby('Fjob').sum()
df2.plot(kind = 'bar', y='studytime')
plt.show()
```

```python
grouped_df = df.groupby(['Fjob','reason']).sum()
grouped_df.unstack().plot(kind ='bar',y='studytime')
plt.show()
```
